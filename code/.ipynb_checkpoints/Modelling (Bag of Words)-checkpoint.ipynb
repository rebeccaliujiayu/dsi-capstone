```python
# Importing general libraries
import numpy as np
import pandas as pd
import re

# Importing visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# Importing libraries for word vectorizing
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
import string

# Importing libraries for model selection and evaluation
from pycaret.classification import *
from sklearn.model_selection import train_test_split
from sklearn import metrics

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
# Downloading stopwords and instantiating lemmatizer
nltk.download('stopwords')
nltk.download("wordnet")
stopwords= nltk.corpus.stopwords.words('english')
wn = nltk.WordNetLemmatizer()
```

```python
# Reading cleaned datasets into Pandas dataframe
ao3 = pd.read_csv("../datasets/clean/ao3_clean.csv")
df = pd.read_csv("../datasets/clean/wiki_clean_sampled.csv")
```

### Preprocessing

I will preprocess the Wikipedia discussion comments by vectorizing them using TfidfVectorizer. PyCaret will be used to run a broad range of classification models on the data, and from there, the best performing model will be optimized and evaluated.

```python
# Creating a function that removes punctuation and numbers, and tokenizes and lemmatizes text
def clean_text(text):
    # Typecast text to string
    text = str(text)
    
    # Remove punctuation and make everything lowercase
    text = "".join([word.lower() for word in text if word not in string.punctuation])   
    
    # Tokenize and lemmatize words
    tokens = re.split('\W+', text)
    text = [wn.lemmatize(word) for word in tokens if word not in stopwords]
    return text
```

```python
# Instantiating tf-idf vectorizer
tfidf_vect = TfidfVectorizer(analyzer=clean_text, min_df = 10)

# Vectorizing train set
X_tfidf = tfidf_vect.fit(df["comment"])
X_train_df = tfidf_vect.transform(df["comment"])
X_train_df = pd.DataFrame(X_train_df.toarray())
X_train_df.columns = tfidf_vect.get_feature_names()

# Putting X_train and y_train into same dataset for PyCaret
train_df = X_train_df.assign(y_attack = df["attack"])
```

```python
# Importing libraries for modelling
model = setup(data = train_df, target = "y_attack")
```

### Modelling
We can see from the results below that the SVM model performed best in terms of accuracy with an accuracy of 83.92% and a precision and recall of 88.72% and 77.78% respectively. It also had a shorter training time compared to runner up models like the logistic regression and extra trees classifier.

```python
best_model = compare_models()
```

After tuning the model, the accuracy of the SVM model increased slightly to 83.99%, with a precision and recall of 78.28% and 88.41% respectively. In the next section, I will evaluate the performance of the tuned SVM model on the holdout set (comprising 4,500 Wikipedia discussion comments), and on the dataset of 320 Ao3 comments.

```python
best_tuned = tune_model(best_model)
```

#### Performance on Ao3 set
Finally, I will test the performance of the tuned SVM model on the 320 Ao3 comments. The performance of the model on the Ao3 comments was much poorer at only 58.75% accuracy, with a precision of 59.33% and a recall of 55.62%. This is likely because the comments that the model was trained on was from a different platform (Wikipedia) from the Ao3 comments.

```python
# Vectorizing test set
X_ao3 = X_tfidf.transform(ao3["comment"])
X_ao3_df = pd.DataFrame(X_ao3.toarray())
X_ao3_df.columns = tfidf_vect.get_feature_names()
```

```python
# Generating predictions for ao3 dataset
svm_tuned_ao3_test = predict_model(best_tuned, data = X_ao3_df)

# Generating confusion matrix
y_ao3_expect = ao3["abusive"]
svm_tuned_ao3_pred = svm_tuned_ao3_test["Label"]
svm_tuned_cf = metrics.confusion_matrix(y_ao3_expect, svm_tuned_ao3_pred)
sns.heatmap(svm_tuned_cf, annot=True, cmap="Blues", fmt = "g")

# Calculating performance metrics
print("Accuracy:", metrics.accuracy_score(y_ao3_expect, svm_tuned_ao3_pred))
print("Precision:", metrics.precision_score(y_ao3_expect, svm_tuned_ao3_pred))
print("Recall:", metrics.recall_score(y_ao3_expect, svm_tuned_ao3_pred))
print("F1 Score:", metrics.f1_score(y_ao3_expect, svm_tuned_ao3_pred))
```

Next, I will examine the composition of the misclassified comments to see where the model is underperforming.

```python
# Subsetting misclassified comments
ao3["pred"] = svm_tuned_ao3_pred
false_neg = ao3[(ao3["abusive"] == 1) & (ao3["pred"]==0)]
false_pos = ao3[(ao3["abusive"] == 0) & (ao3["pred"]==1)]

# Instantiating tf-idf vectorizer
test_vect = TfidfVectorizer(analyzer=clean_text)

# Vectorizing abusive comments
fneg_tfidf = test_vect.fit_transform(false_neg["comment"])
fneg_df = pd.DataFrame(fneg_tfidf.toarray())
fneg_df.columns = test_vect.get_feature_names()

# Vectorizing non-abusive comments
fpos_tfidf = test_vect.fit_transform(false_pos["comment"])
fpos_df = pd.DataFrame(fpos_tfidf.toarray())
fpos_df.columns = test_vect.get_feature_names()
```

```python
# Removing common words -- these are all character names
del_words = ["wwx", "wei", "wuxian", "ying",
             "lwj", "lan", "wangji", "zhan",
             "jc", "jiang", "cheng",
             "jgy", "jin", "guangyao", "meng", "yao",
             "lxc", "xichen",
             "jfm", "jiang", "fengmian",
             "wangxian", "xiyao", "sangcheng",
             "story", "mdzs"]

for word in del_words:
    if word in fpos_df.columns:
        fpos_df.drop(columns = word, inplace = True)
    if word in fneg_df.columns:
        fneg_df.drop(columns = word, inplace = True)

# Converting to dictionary as wordcloud requires frequencies in dictionary format
fpos_del = fpos_df.sum(axis = 0).to_dict()
fneg_del = fneg_df.sum(axis = 0).to_dict()
```

From the word clouds below, it's apparent that the bag of word model tends to misclassify abusive comments that don't have profanity in it, but might have criticisms that are specific to fanfiction, like "ooc" which is a common criticism that stands for "out of character". On the other hand, it tends to misclassify positive comments that have profanity in it.

```python
# Plotting wordcloud for flagged comments
fneg_cloud = WordCloud(width = 800, height = 800, background_color ='white', colormap = "plasma",
                          stopwords = stopwords, min_font_size = 10).generate_from_frequencies(fneg_del)

plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
plt.imshow(fneg_cloud)
plt.axis("off")
plt.title("Top Words (False Negatives)");

# Plotting wordcloud for unflagged comments
fpos_cloud = WordCloud(width = 800, height = 800, background_color ='white', colormap = "viridis",
                stopwords = stopwords, min_font_size = 10).generate_from_frequencies(fpos_del)

plt.subplot(1,2,2)
plt.imshow(fpos_cloud)
plt.axis("off")
plt.title("Top Words (False Positives)");
```

Seen below are examples of positive comments that have been misclassified as negative comments. These include (1) positive comments which contain profanity like "oh fuck it's going down and i'm not ready" and (2) positive comments in which readers speak negatively about antagonists, such as "(antagonist name) is a fucking bastard."

As compared to bag of word models, which do not take sequence of words into account, deep learning models may perform better when it comes to misclassification of comments that contain profanity. However, without labelled training data that is more tailored to our purpose, deep learning models will probably be unable to learn to differentiate comments that are negative towards an antagonist, from comments that are negative towards the author.

As predicted, the classifier performed better on trolling styles that were highly toxic and utilized profane and violent language, or hate speech. In comparison, it did more poorly on trolls who tended to use more fandom-specific attacks, or who were more sarcastic.
